# Categorize Loan Purpose:

```python
def Categorizer(Purpose):
    if Purpose in ['debt_consolidation','credit_card']:
        return 'Financial'
    elif Purpose in ['small_business','educational']:
        return 'Educational/Business'
    else:
        return 'Other'
```

```python
merged_data['purpose_category'] = merged_data['purpose'].apply(Categorizer)
merged_data
```

# Asses the Risk
### dti>20 and delinq.2yrs > 2 and revol.util>60 >>>>> is high risk borrower
### else is low Risk borrower

```python
def asses_risk(row):
    if row['dti']>20 and row['delinq.2yrs'] > 2 and row['revol.util']>60:
        return 'High Risk'
    else:
        return 'Low Risk'
```

```python
merged_data['Risk'] = merged_data.apply(asses_risk, axis=1)
merged_data
```

# Asses FICO score

```python
def fico_score_categorizer(fico_score):
    if fico_score > 800 and fico_score <= 850:
        return 'Excellent'
    elif fico_score >= 740 and fico_score < 800:
        return 'Very Good'
    elif fico_score >= 670 and fico_score < 740:
        return 'Good'
    elif fico_score >= 580 and fico_score < 670:
        return 'Fair'
    else:
        return 'Poor'
```

```python
merged_data['fico_category'] = merged_data['fico'].apply(fico_score_categorizer)
merged_data
```

# Identify customers with more than avg inquires and derogatory records

```python
def identify_high_inq_deg(row):
    
    avg_inq = merged_data['inq.last.6mths'].mean()
    avg_derog = merged_data['pub.rec'].mean()
    
    if row['inq.last.6mths']> avg_inq and row['pub.rec'] > avg_derog:
        return True
    else:
        return False
```

```python
merged_data['High_Inquires_and_Public_Records'] = merged_data.apply(identify_high_inq_deg, axis=1)
merged_data
```
